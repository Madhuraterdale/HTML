Decorators

```python
def add_by(function):
  def  wrapper(*args):
    result=function(*args)+10
    return result
  return wrapper
```

```python
@add_by
def summ(a: int , b: int)->int:
  result= a+b
  return result
@add_by
def sub(a: int , b: int)->int:
  result= a-b
  return result
@add_by
def mul(a: int , b: int)->int:
  result= a*b
  return result
@add_by
def div(a: int , b: int)->int:
  result= a/b
  return result
```

```python
summ(10,23)
```
